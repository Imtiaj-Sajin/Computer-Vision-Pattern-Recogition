```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[2,9,2,98,4]

plt.plot(x,y)
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(x,y)
```

```python
plt.plot(x,y, color='green')
```

```python
plt.plot(x,y,color='magenta',linewidth='4')
```

```python
plt.plot(x,y,color='red',linewidth='1',marker='+')
```

```python
plt.plot(x,y,color='red',linewidth='1',marker='+',linestyle=':')
```

```python
plt.plot(x,y,'b+--') #blue marker='+' linestyle='--' dashed line
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

```python
plt.plot(x,y,'g*',alpha=0.5) #g=green color,marker=*, alpha=scale
```

# **figure()**

```python
import numpy as np
y1=np.sin(x)
y2=np.cos(x)
```

```python
plt.figure()
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(('sin','cos'))
plt.show()
```

```python
x=np.arange(-10,10,.01)
y1=np.sin(x)
y2=np.cos(x)

plt.figure()
plt.plot(x,y1,x,y2)
plt.legend(['sin','cos'])
plt.show()
```

# **Subplot**

```python
plt.figure(figsize=(20,5))
plt.subplot(1,2,1) #1,2,1 mean 1 row, 2 columns, and this is the first subplot.
plt.plot(x,y1)
plt.legend('Sine')

plt.subplot(1,2,2) #1,2,2 mean 1 row, 2 columns, and this is the second subplot.
plt.plot(x,y2)
plt.title('Cosine')

plt.show()
```
