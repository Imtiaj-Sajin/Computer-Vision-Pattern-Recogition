```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5]
y=[2,9,2,98,4]

plt.plot(x,y)
```

```python
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.plot(x,y)
```

```python
plt.plot(x,y, color='green')
```

```python
plt.plot(x,y,color='magenta',linewidth='4')
```

```python
plt.plot(x,y,color='red',linewidth='1',marker='+')
```

```python
plt.plot(x,y,color='red',linewidth='1',marker='+',linestyle=':')
```

```python
plt.plot(x,y,'b+--') #blue marker='+' linestyle='--' dashed line
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```
